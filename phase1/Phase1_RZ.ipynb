### **1. What is language modeling?**

Language modeling is the task of estimating the probability of the following token sequences given previous token sequences.

---

### **2. What is self-supervised pretraining?**

Self-supervised pretraining is to train the model using the supervision signals generated by itself, rather than those provided by humans. This is generally done by constructing its own training tasks directly from unlabeled data, such as having the system create pseudo labels. It’s the dominant paradigm for LLMs due to scalability on unlabeled corpora.

---

### **3. Why is pretraining more hardware-efficient for Transformer- or attention-based models compared to RNN-based models?**

- **RNNs** process tokens sequentially; computation can’t be parallelized across sequence positions.
- **Transformers** use self-attention, allowing **parallel computation across the entire sequence**.
- This makes training on GPUs (and TPUs) more efficient and hardware-friendly.

---

### **4. What is the difference between encoder-only and decoder-only models, and why are decoder-only models more popular?**

- **Encoder-only models** use the entire sequence as input and the give the encode their information into a real-valued vector. Because the encoder does not work as a standalone NLP system, it is often integrated as a component into a bigger system.
- **Decoder-only models** use the past tokens as input and generate the prediction of probability distribution of the following tokens. 
- The decoder-only architecture has been widely used in developing language models since it can be used to predict token probabilities. For example, we can use a Transformer decoder as a language model by simply removing cross-attention sub-layers from it. Such a model predicts the distribution of tokens at a position given its preceding tokens, and the output is the token with the maximum probability.

### **5. Log cross-entropy loss calculation**

$
\text{Loss} = -\log(\text{Predicted probability of correct word})
$

- If correct word is **Cherry** (3rd):
  $\text{Loss} = -\log(0.2) \approx 1.6094$

- If correct word is **Banana** (2nd):
  $\text{Loss} = -\log(0.7) \approx 0.3567$

### **6. What are zero-shot learning, few-shot learning, and in-context learning?**

- **Zero-shot learning**: Model performs a task from instructions without seeing any examples.
- **Few-shot learning**: Prompt includes a few input–output examples.
- **In-context learning**: prompts involve demonstrations of problem-solving, and LLMs can learn from these demonstrations how to solve new problems. No weight changes are involved.

---

### **7. Why is fine-tuning necessary? What is instruction fine-tuning?**

- While prompting is a powerful way to adapt large language models, some tuning efforts are still needed to ensure the models can follow instructions accurately. The **fine-tuning** process is crucial for **aligning the values of these models with human values**. The token prediction ability learned in the pre-training phase can be generalized to accomplish new tasks. The idea behind **fine-tuning** is that some general knowledge of language has been acquired in pre-training, but we need a mechanism to **activate this knowledge for applying it to new tasks**.
- **Instruction fine-tuning**: Model is trained on pairs of **instructions + responses** so it learns to follow diverse natural language prompts.

---

### **8. What is tokenization? What is a word embedding layer?**

- **Tokenization**: Splits text into units (tokens). Most LLMs use **subword units** (e.g., BPE or SentencePiece).
- **Word Embedding layer**: Maps discrete tokens (words) into low-dimensional real-valued vectors, making it possible to compute the meanings of words in a continuous space.

---

### **9. What is position embedding? What kind of position embedding method is used in Llama models?**

- **Position embeddings**: add information about token order into the tokens.
- **LLama uses rotary position embeddings**: Applies sinusoidal rotation during the computation of attention queries and keys.

---

### **10. What is the difference between multi-head attention and grouped-query attention? Which type is used in Llama?**

- **Multi-head attention (MHA):** Each head computes its own Q/K/V, enabling diverse representations.
- **Grouped-query attention (GQA):** Shares keys/values across multiple heads; fewer KV projections.
- **Differences:** **LLaMA uses GQA**, since it reduces memory and speeds up inference.

---

### **11. What kind of activation function is used in LLaMA models?**

**LLaMA uses SwiGLU**, a combination of Swish activation and GLU activation:

$
\text{SwiGLU}(x, W, V, b, c\beta) = \text{Swish}_\beta(xW+b) \otimes (xV+c)
$


---

### **12. What is layer normalization? What’s the difference from batch normalization?**

- **LayerNorm**: Normalizes across features within a single sample in one layer.
- **BatchNorm**: Normalizes across the batch for one layer.

**LayerNorm is used in Transformers** because it's position-invariant and works well with variable-length sequences.

### **13. What is the auto-regressive generation process, and how is a decoding strategy used during text generation?**

- **Auto-regressive generation**: predict one token at a time, append it to the context to predict the next token.

- **Decoding strategies** is used to pick up the next token among the predicted probability distribution of all the tokens, usually the one with the highest-probability is chosen.
